# 01 — Exploratory Data Analysis

Explore the raw rental market datasets:
- Zillow Observed Rent Index (ZORI)
- Zillow Home Value Index (ZHVI)
- Redfin Median Asking Rent

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

RAW = Path("../data/raw")
```

## 1. Zillow Observed Rent Index (ZORI)

```python
zori = pd.read_csv(RAW / "zillow_observed_rent_index.csv")
print(zori.shape)
zori.head()
```

```python
zori.info()
zori.describe()
```

## 2. Zillow Home Value Index (ZHVI)

```python
zhvi = pd.read_csv(RAW / "zillow_home_value_index.csv")
print(zhvi.shape)
zhvi.head()
```

```python
zhvi.info()
zhvi.describe()
```

## 3. Redfin Median Asking Rent

```python
redfin = pd.read_csv(
    RAW / "redfin_median_asking_rent.csv",
    encoding="utf-16",
    sep="\t"
)
print(redfin.shape)
redfin.head()
```

```python
redfin.info()
redfin.describe()
```
